```python
def clean_text(cleaned_texts,out_text):
    for text in out_text:
        text = text.replace("\n","")
        text = text.replace("..","")
        text = text.replace("  ","")
        text = text.replace("\x0c","")
        for i in range(50):
            text = text.replace("\n{}".format(i),"")
        if (str(text)!=' ') and (str(text)!=''):
            cleaned_texts.append(text)
```
