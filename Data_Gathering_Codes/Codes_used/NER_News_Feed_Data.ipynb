```python
# pip install -U spacy
# python -m spacy download en_core_web_sm
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob
import pandas as pd
import os
```

```python
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')
```

```python
# Process whole documents
# Open the text file in read mode
with open('news_feed/article_contents_cleaned.txt', 'r') as file:
    
    # Read the contents of the file into a variable
    file_contents = file.read()
#make a doc object    
doc = nlp(file_contents)
```

```python
file_contents
```

```python
keyword = "technology"

# Tokenize the keyword using spaCy
doc_keyword = nlp(keyword)

# Iterate over the sentences in the data and calculate the similarity score for each sentence and the keyword
sentences = []
similarities = []
polarities = []
subjectivities = []
for sentence in doc.sents:
    similarity = sentence.similarity(doc_keyword)
    polarity = sentence._.blob.polarity
    subjectivity = sentence._.blob.subjectivity
    sentences.append(sentence.text)
    similarities.append(similarity)
    polarities.append(polarity)
    subjectivities.append(subjectivity)
```

```python
df = pd.DataFrame({"sentence": sentences, "similarity": similarities,"polarities":polarities,"subjectivities":subjectivities})
```

```python
df["polarities"] = df["polarities"].round(2)
```

```python
df = df.sort_values(by="polarities", ascending=False)
```

```python
df=df.reset_index().drop(columns=["index"])
```

```python
df_final = df[df["similarity"]>0.2]
```

```python
df_final
```

```python
df_final.to_csv("article_contents.csv")
```
